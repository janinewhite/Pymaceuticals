```python
#import dependencies and data
import pandas as pd
import matplotlib.pyplot as plt
import statistics as st
colors = ["red","green","blue","orange","blueviolet","gray","cyan","darkgoldenrod","darkorchid","deeppink"]
markers = ["o","v","^","<",">","8","s","*","D","+"]
```

```python
#Read tumor data
tumor_file = "Pymaceuticals/data/clinicaltrial_data.csv"
tumor_data = pd.read_csv(tumor_file)
tumor_data.head()
```

```python
#Read drug data
drug_file = "Pymaceuticals/data/mouse_drug_data.csv"
drug_data = pd.read_csv(drug_file)
drug_data.head()
```

```python
mouse_IDs = drug_data["Mouse ID"].unique()
print(mouse_IDs)
drugs = drug_data["Drug"].unique()
num_drugs = len(drugs)
print(drugs)
```

```python
#Merge files
mice_data = tumor_data.merge(drug_data, on="Mouse ID")
mice_data.head()
```

```python
#Tumor size and Metastatic sites trend by Drug
visualization_index = 0
for drug in drugs:
    figs = plt.figure(figsize=(8,4))
    ax0 = figs.add_subplot(121)
    ax0.set_ylim([0,80])
    ax1 = figs.add_subplot(122)
    ax1.set_ylim([0,5])
    series = mice_data.loc[mice_data["Drug"]==drug]
    series.plot.scatter(x='Timepoint',y='Tumor Volume (mm3)',c=colors[visualization_index],marker=markers[visualization_index],label=drug,ax=ax0)
    series.plot.scatter(x='Timepoint',y='Metastatic Sites',c=colors[visualization_index],marker=markers[visualization_index],label=drug,ax=ax1)
    visualization_index += 1
```
